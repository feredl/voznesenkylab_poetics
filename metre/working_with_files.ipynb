```python
import ru_accent_poet
from pathlib import Path
import shutil
import re
import pandas as pd
from transliterate import translit
from number_to_string import get_string_by_number
```

```python
voz_poems = pd.read_excel("Вознесенский для ЦВ (1) (1).xlsx")
voz_poems = pd.DataFrame(voz_poems)
voz_poems["name"] = list(range(0, voz_poems.shape[0]))
voz_poems
```

```python
for i in voz_poems["n"]:
    with open(f"{i}.txt", "w+", encoding = "utf-8") as f:
        poem = voz_poems["Текст"].iloc[i]
        poem = translit(poem, "ru")
        f.write(poem)
    i += 1
```

```python
data_dir = Path("voz.ipynb").parent
output_dir = data_dir / "poems"
output_dir.mkdir(exist_ok = True)
```

```python
for file_path in data_dir.glob("*.txt"):
    file_name = file_path.name
    just_name, no_need = file_name.rsplit(".", maxsplit=1)
    try:
        just_name = int(just_name)
    except:
        continue
    a = list(range(0, 1340))
    if just_name in a:
        ru_accent_poet.write_file([file_name])
```

```python
for file_path in data_dir.glob("*accented.txt"):
    shutil.move(file_path, output_dir)
```

```python
voz_poems_new = pd.read_csv("METRES.csv")
voz_poems_new = pd.DataFrame(voz_poems_new)
voz_poems_new = voz_poems_new.merge(voz_poems, on = ["name"])
voz_df = pd.DataFrame({"name": voz_poems_new["Название"], "number": voz_poems_new["name"], "meter": voz_poems_new["meter_poem"], "accents": voz_poems_new["accents"], "poem": voz_poems_new["Текст"]})
voz_df.to_excel("voz_metres.xlsx")
```
